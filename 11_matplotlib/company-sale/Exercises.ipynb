# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
df = pd.read_csv(url)
df = df.set_index('month_number')
df
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
import matplotlib.pyplot as plt

data = df['total_profit']
plt.figure(figsize=(8,5), dpi=100)
plt.plot(data)
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.xticks(range(1, 13))
plt.yticks(range(100000, 600000, 100000))
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
import matplotlib.pyplot as plt

data = df['total_profit']

plt.figure(figsize=(8, 5), dpi=100)

plt.plot(data, label='Profit data of last year', linestyle='--', linewidth=3.5, marker='o', markersize=9, markerfacecolor='black', markeredgecolor='red', color="red")

plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company Sale data of last year')
plt.xticks(range(1, 13))
plt.yticks(range(100000, 600000, 100000))

plt.legend(loc='lower right', fontsize=13)

plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5), dpi=100)

products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

for product in products:
    plt.plot(df[product], label=f'{product.title()} Sales Data', linewidth=4, marker='o', markersize=8)

plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.xticks(range(1, 13))
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.legend(fontsize=12)
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
import matplotlib.pyplot as plt

data = df['toothpaste']
plt.figure(figsize=(8,5), dpi=100)
plt.scatter(range(1,13), data, label='ToothPaste Sales Data', s=50, marker='o', zorder=1)
#plt.plot(range(1, 13), data, linestyle='-', linewidth=1.5, alpha=0.7, zorder=1)
plt.grid(True, linestyle='--', linewidth=1.5)
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('ToothPaste Sales data')
plt.xticks(range(1, 13))
plt.legend(fontsize=12)
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
cream = df['facecream']
wash = df['facewash']
plt.figure(figsize=(8,5), dpi=100)
plt.bar(df.index-0.13, cream, width=0.25)
plt.bar(df.index+0.13, wash, width=0.25)
plt.grid(True, linestyle='--', linewidth=1.5)
plt.xticks(range(1, 13))
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Facewash and facecream sales data')
plt.legend(labels=["Face Cream sales data", "Face Wash sales data"], loc="upper left", fontsize=13)
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](img/06.png)

```python
data = df['bathingsoap']
plt.figure(figsize=(8,5), dpi=100)
plt.bar(df.index, data, width=0.75)
plt.grid(True, linestyle='--', linewidth=1.5)
plt.xticks(range(1, 13))
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Bathingsoap sales data')
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
data = df['total_profit']
plt.figure(figsize=(8,5), dpi=100)
plt.hist(data, bins=[150000, 175000, 200000, 225000, 250000, 300000, 350000])
plt.yticks(range(0, 7))
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.title('Profit data')
plt.legend(labels=['Profit data'], loc='upper left', fontsize=11)
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
langs = ['FaceCream', 'FaceWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
check = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
Sales = []
for colum in check:
    Sales.append(df[colum].sum())

plt.figure(figsize=(5, 5))
plt.pie(Sales, labels=langs, autopct='%1.1f%%')
plt.legend(labels=langs, loc='lower right', fontsize=10)
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](img/09.png)

```python
bathingsoap = df['bathingsoap']
facewash = df['facewash']

plt.figure(figsize=(8,4), dpi=1000)
ax1 = plt.subplot(211)
ax1.plot(bathingsoap, linewidth=3, marker='o', markersize=8, color="black")
ax1.set_title('Sales data of a Bathingsoap')
ax1.set_xticks(list(range(1, 13)))
ax1.set_yticks([7500, 10000, 12500])

ax2 = plt.subplot(212, sharex=ax1)
ax2.plot(facewash, linewidth=3, marker='o', markersize=8, color="red")
ax2.set_title('Sales data of a facewash')
ax2.set_yticks([1500, 2000])

plt.xlabel("Month Number")
plt.ylabel('sales units in number')
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
plt.figure(figsize=(8,4), dpi=1000)
data = [df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer']]
colors = ['m', 'c', 'r', 'k', 'g', 'y']
plt.stackplot(df.index, data, labels=df.columns, colors=colors)
plt.title('All product sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.legend(loc='upper left')
plt.show()
```
